```python
def translate_to_telugu(word, verbose=False):
    try:
        translated = translator.translate(word, src='en', dest='te')
        if verbose:
            print(f"English: {word} → Telugu: {translated.text}")
        return translated.text
    except Exception as e:
        print(f"Translation Error for {word}: {e}")
        return word
```

```python
import re
from indicnlp.tokenize import indic_tokenize


print_counter = 0
max_prints = 5

def preprocess_text(text):
    global print_counter
    try:

        original_text = text
        text = re.sub(r'[^\u0C00-\u0C7F\s]', '', text)
        text = re.sub(r'\d+', '', text)
        tokens = indic_tokenize.trivial_tokenize(text)
        translated_tokens = [translate_to_telugu(word) if re.search('[a-zA-Z]', word) else word for word in tokens]
        tokens = [word for word in translated_tokens if word not in telugu_stopwords]

        # Print only for the first 5 samples
        if print_counter < max_prints:
            print(f"\nSample {print_counter+1}:")
            print(f"Original Text: {original_text}")
            print(f"After Removing Special Characters & Numbers: {text}")
            print(f"Tokenized Words: {tokens}")
            print(f"After Translating English Words: {translated_tokens}")
            print(f"After Removing Stopwords: {tokens}")
            print_counter += 1

        return " ".join(tokens)

    except Exception as e:
        print(f"Error in Preprocessing: {e}")
        return ""


df['cleaned_text'] = df['text'].apply(preprocess_text)


print("\nSample Preprocessed Text (Showing 5 Rows Only):")
print(df[['text', 'cleaned_text']].head(5))
```

```python
import gensim
import numpy as np

# Load FastText model
fasttext_model = gensim.models.KeyedVectors.load_word2vec_format('cc.te.300.vec', binary=False)

def get_embedding(word):
    return fasttext_model[word] if word in fasttext_model else np.random.uniform(-0.01, 0.01, 300)

def compute_sentence_embedding(text):
    words = text.split()
    word_vectors = [get_embedding(w) for w in words]
    return np.mean(word_vectors, axis=0) if word_vectors else np.random.uniform(-0.01, 0.01, 300)

df['fasttext_embeddings'] = df['cleaned_text'].apply(compute_sentence_embedding)
X_fasttext = np.stack(df['fasttext_embeddings'].values)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import numpy as np

# Load IndicBERT sentiment model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-xlm-roberta-base-sentiment")
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-xlm-roberta-base-sentiment")

labels = ['negative', 'neutral', 'positive']


def get_telugu_sentiment_score(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        logits = model(**inputs).logits
    probs = torch.softmax(logits, dim=1).numpy()[0]
    sentiment_label = labels[np.argmax(probs)]
    sentiment_score = probs[np.argmax(probs)]


    polarity_map = {'positive': 1, 'neutral': 0, 'negative': -1}
    polarity = polarity_map[sentiment_label] * sentiment_score

    return polarity


df['sentiment'] = df['cleaned_text'].apply(get_telugu_sentiment_score)


print(df[['cleaned_text', 'sentiment']].head())
```

```python
def lexical_features(text):
    tokens = text.split()
    num_tokens = len(tokens)
    num_unique_tokens = len(set(tokens))
    avg_word_len = np.mean([len(word) for word in tokens]) if tokens else 0
    vocab_richness = num_unique_tokens / num_tokens if num_tokens > 0 else 0

    return [avg_word_len, vocab_richness]

lexical_feats = np.array([lexical_features(t) for t in df['cleaned_text']])
```

```python
import stanza
import numpy as np
import pandas as pd

nlp = stanza.Pipeline(lang='te', processors='tokenize,pos', use_gpu=True)



def pos_features(text):
    doc = nlp(text)

    pos_counts = {}
    total_tokens = 0

    for sentence in doc.sentences:
        for word in sentence.words:
            total_tokens += 1
            pos = word.upos  # Universal POS tags
            pos_counts[pos] = pos_counts.get(pos, 0) + 1

    # List of possible POS tags based on Universal POS tagset
    pos_tags = ['NOUN', 'VERB', 'ADJ', 'ADV', 'PROPN', 'PRON', 'AUX', 'ADP', 'CONJ', 'DET', 'NUM', 'PART', 'INTJ', 'PUNCT', 'SYM', 'X']

    # Compute ratios for each POS tag
    pos_ratios = [pos_counts.get(tag, 0) / total_tokens if total_tokens > 0 else 0 for tag in pos_tags]

    # Derived features
    noun_count = pos_counts.get('NOUN', 0)
    verb_count = pos_counts.get('VERB', 0)
    adj_count = pos_counts.get('ADJ', 0)
    propn_count = pos_counts.get('PROPN', 0)
    punct_count = pos_counts.get('PUNCT', 0)

    noun_verb_ratio = noun_count / verb_count if verb_count > 0 else 0
    adj_density = adj_count / total_tokens if total_tokens > 0 else 0
    propn_ratio = propn_count / total_tokens if total_tokens > 0 else 0
    punct_ratio = punct_count / total_tokens if total_tokens > 0 else 0


    return pos_ratios + [noun_verb_ratio, adj_density, propn_ratio, punct_ratio]

pos_feats = np.array([pos_features(t) for t in df['cleaned_text']])


pos_tags = ['NOUN', 'VERB', 'ADJ', 'ADV', 'PROPN', 'PRON', 'AUX', 'ADP', 'CONJ', 'DET', 'NUM', 'PART', 'INTJ', 'PUNCT', 'SYM', 'X']
feature_names = pos_tags + ['noun_verb_ratio', 'adj_density', 'propn_ratio', 'punct_ratio']


pos_feats_df = pd.DataFrame(pos_feats, columns=feature_names)


df_with_feats = pd.concat([df, pos_feats_df], axis=1)


print(df_with_feats.head())
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import keras_tuner as kt



X = final_features
y = df['label'].values
num_classes = len(np.unique(y))
y = to_categorical(y, num_classes=num_classes)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
input_dim = X_train.shape[1]


def build_model(hp):
    model = Sequential()
    model.add(Dense(
        units=hp.Choice('dense1_units', [256, 512, 768]),
        activation='relu',
        input_shape=(input_dim,),
        kernel_regularizer=l2(hp.Choice('l2_1', [0.001, 0.005, 0.01]))
    ))
    model.add(Dropout(hp.Float('dropout1', 0.2, 0.5, step=0.1)))

    model.add(Dense(
        units=hp.Choice('dense2_units', [128, 256, 512]),
        activation='relu',
        kernel_regularizer=l2(hp.Choice('l2_2', [0.001, 0.005, 0.01]))
    ))
    model.add(Dropout(hp.Float('dropout2', 0.2, 0.5, step=0.1)))

    model.add(Dense(
        units=hp.Choice('dense3_units', [64, 128, 256]),
        activation='relu',
        kernel_regularizer=l2(hp.Choice('l2_3', [0.001, 0.005, 0.01]))
    ))
    model.add(Dropout(hp.Float('dropout3', 0.2, 0.5, step=0.1)))

    model.add(Dense(num_classes, activation='softmax'))

    model.compile(
        optimizer=Adam(learning_rate=hp.Choice('lr', [1e-2, 1e-3, 5e-4, 1e-4])),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model


tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=15,
    executions_per_trial=1,
    directory='ffnn_tuner',
    project_name='self_tuning_telugu_aggression'
)

callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', patience=5, factor=0.5, verbose=1)
]


tuner.search(
    X_train, y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)


best_model = tuner.get_best_models(num_models=1)[0]

history = best_model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)


y_pred = np.argmax(best_model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)
print(classification_report(y_true, y_pred))


def plot_training_history(history):
    sns.set(style="whitegrid")
    plt.figure(figsize=(14, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy', color='blue')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training & Validation Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss', color='blue')
    plt.plot(history.history['val_loss'], label='Validation Loss', color='red')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training & Validation Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()


def plot_metrics(y_true, y_pred, num_classes):
    report = classification_report(y_true, y_pred, output_dict=True)
    classes = [str(i) for i in range(num_classes)]

    precision = [report[c]['precision'] for c in classes]
    recall = [report[c]['recall'] for c in classes]
    f1 = [report[c]['f1-score'] for c in classes]

    plt.figure(figsize=(10, 5))
    x = np.arange(len(classes))
    width = 0.2

    plt.bar(x - width, precision, width, label='Precision', color='skyblue')
    plt.bar(x, recall, width, label='Recall', color='salmon')
    plt.bar(x + width, f1, width, label='F1-Score', color='lightgreen')

    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Precision, Recall, and F1-Score per Class')
    plt.xticks(x, classes)
    plt.legend()
    plt.show()


plot_training_history(history)
plot_metrics(y_true, y_pred, num_classes)
```

```python
import gensim
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Bidirectional, Dense, Dropout, Layer, BatchNormalization, Lambda
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Sample Data
# final_features = np.random.rand(1000, 300)  # Example feature shape (1000 samples, 300 features)
# df = pd.DataFrame({'label': np.random.randint(0, 3, 1000)})

X = final_features
y = df['label'].values

# One-hot encode labels
num_classes = len(np.unique(y))
y = to_categorical(y, num_classes=num_classes)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Reshape for LSTM (samples, timesteps, features) — here timesteps=1
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Compute Class Weights
class_weights = compute_class_weight('balanced', classes=np.unique(np.argmax(y, axis=1)), y=np.argmax(y, axis=1))
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}

# Attention Layer
class AttentionLayer(Layer):
    def __init__(self, trainable=True):
        super(AttentionLayer, self).__init__()
        self.trainable = trainable

    def build(self, input_shape):
        self.W = self.add_weight(shape=(input_shape[-1], input_shape[-1]),
                                 initializer="random_normal",
                                 trainable=self.trainable)
        self.b = self.add_weight(shape=(input_shape[-1],),
                                 initializer="zeros",
                                 trainable=self.trainable)

    def call(self, inputs):
        score = tf.matmul(inputs, self.W) + self.b
        attention_weights = tf.nn.softmax(score, axis=1)
        return tf.reduce_sum(inputs * attention_weights, axis=1)

# Model
input_layer = Input(shape=(X_train.shape[1], X_train.shape[2]))

bilstm = Bidirectional(LSTM(128, return_sequences=True, dropout=0.3, recurrent_dropout=0.3))(input_layer)
batch_norm = BatchNormalization()(bilstm)
attention1 = AttentionLayer(trainable=True)(batch_norm)
attention1 = Lambda(lambda x: tf.expand_dims(x, axis=1))(attention1)

bilstm2 = Bidirectional(LSTM(64, return_sequences=True, dropout=0.3, recurrent_dropout=0.3))(attention1)
batch_norm2 = BatchNormalization()(bilstm2)
attention2 = AttentionLayer(trainable=True)(batch_norm2)

dropout = Dropout(0.3)(attention2)
dense1 = Dense(64, activation='relu', kernel_regularizer=l2(0.005))(dropout)
dropout2 = Dropout(0.3)(dense1)
dense2 = Dense(32, activation='relu', kernel_regularizer=l2(0.005))(dropout2)
output_layer = Dense(num_classes, activation='softmax')(dense2)

# Compile model
optimizer = Adam(learning_rate=0.001, clipnorm=1.0)
model = Model(inputs=input_layer, outputs=output_layer)
model.compile(loss=tf.keras.losses.CategoricalCrossentropy(), optimizer=optimizer, metrics=['accuracy'])

# Callbacks
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train Model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    verbose=1,
    callbacks=[lr_scheduler, early_stopping],
    class_weight=class_weight_dict
)

# Evaluate the model
y_pred = np.argmax(model.predict(X_test), axis=1)
y_true = np.argmax(y_test, axis=1)
print(classification_report(y_true, y_pred))

# Plot Training History
def plot_training_history(history):
    sns.set(style="whitegrid")
    plt.figure(figsize=(14, 5))

    # Accuracy Plot
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy', color='blue')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training & Validation Accuracy')
    plt.legend()

    # Loss Plot
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss', color='blue')
    plt.plot(history.history['val_loss'], label='Validation Loss', color='red')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training & Validation Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Plot Precision, Recall, F1-Score per Class
def plot_metrics(y_true, y_pred, num_classes):
    report = classification_report(y_true, y_pred, output_dict=True)
    classes = [str(i) for i in range(num_classes)]

    precision = [report[c]['precision'] for c in classes]
    recall = [report[c]['recall'] for c in classes]
    f1 = [report[c]['f1-score'] for c in classes]

    plt.figure(figsize=(10, 5))
    x = np.arange(len(classes))
    width = 0.2

    plt.bar(x - width, precision, width, label='Precision', color='skyblue')
    plt.bar(x, recall, width, label='Recall', color='salmon')
    plt.bar(x + width, f1, width, label='F1-Score', color='lightgreen')

    plt.xlabel('Class')
    plt.ylabel('Score')
    plt.title('Precision, Recall, and F1-Score per Class')
    plt.xticks(x, classes)
    plt.legend()
    plt.show()

# Generate Plots
plot_training_history(history)
plot_metrics(y_true, y_pred, num_classes)
```
